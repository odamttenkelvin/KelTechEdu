```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David'],
    'Attendance (%)': [95, 80, 60, 45]
})

df.plot(x='Student', y='Attendance (%)', kind='bar', title='Attendance Overview')
plt.axhline(y=75, color='r', linestyle='--')  # Risk threshold
plt.show()
```
